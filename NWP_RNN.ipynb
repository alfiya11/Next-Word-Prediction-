```python
import numpy as np
import pandas as pd
import csv
from numpy import array
from keras.preprocessing.text import Tokenizer
from keras.utils import to_categorical
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Embedding
import tensorflow as tf
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
df = pd.read_csv('/content/tmdb_5000_movies.csv')
df
```

```python
df = df['original_title']
df
```

```python
movie_name = df.to_list()
movie_name
```

```python
# integer encode text

tokenizer = Tokenizer()
tokenizer.fit_on_texts([movie_name])
encoded_data = tokenizer.texts_to_sequences([movie_name])[0]
encoded_data
```

```python
#deterine vocabulary size
vocab_size = len(tokenizer.word_index) + 1
print("vocab size: %d" % vocab_size)
```

```python
#creating word -> word sequences

sequences = list()
for i in range(1, len(encoded_data)):
  sequence = encoded_data[i-1:i+1]
  sequences.append(sequence)
print("Total Sequences: %d" % len(sequences))
# split int x & y
```

```python
# we have 4802 pairs of i/p and o/p to train n/w
```

```python
sequences
```

```python
# dividing seq into i/p & o/p
sequences = array(sequences)
X, y = sequences[:,0], sequences[:,1]
```

```python
X[:5]
```

```python
y[:5]
```

```python
# one hot encode o/p
y = to_categorical(y, num_classes = vocab_size)
y[:5]
```

```python
model = Sequential()
model.add(Embedding(vocab_size, 10, input_length = 1))
model.add(LSTM(50))
model.add(Dense(vocab_size, activation = 'softmax'))
print(model.summary())
```

```python
#compile n/w
model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])

#fit n/w
model.fit(X, y, epochs = 200)
```

```python
def generate_seq(model, tokenizer, enter_text, n_pred):
	in_text, result = enter_text, enter_text  #
	# generate a fixed number of words
	for _ in range(n_pred):

		# encode the text as integer
		encoded = tokenizer.texts_to_sequences([in_text])[0]
		encoded = array(encoded)

		# predict a word in the vocabulary
		yhat = tf.config.run_functions_eagerly(True)

		# map predicted word index to word
		out_word = ''
		for word, index in tokenizer.word_index.items():
			if index == yhat:
				out_word = word
				break
		# append to input
		in_text, result = out_word, result + ' ' + out_word
	return result
```
